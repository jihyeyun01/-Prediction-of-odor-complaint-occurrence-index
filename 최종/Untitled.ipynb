```python
b_df = pd.read_csv("B관측소_시간별_민원발생건수.csv", encoding='euc-kr')
b_df
```

```python
c_df = pd.read_csv("C관측소_시간별_민원발생건수.csv", encoding='euc-kr')
c_df
```

```python
total_df = pd.concat([b_df, c_df])
total_df = total_df.reset_index()
total_df = total_df.drop(['index','NO'], axis = 1)
```

```python
total_df['월일'] = total_df['발생일자'].astype(str).str[5:11]
```

```python
total_df.sort_values(by='발생건수',ascending=False)
```

```python
total_df.groupby(['월일']).sum().plot.bar()
```
