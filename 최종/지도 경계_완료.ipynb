```python
import folium
```

```python
import geopandas as gpd
sig = gpd.read_file('emd.shp', encoding= 'cp949')
sig
```

```python
sig_1 = sig[(sig['EMD_KOR_NM']=='대화동')|(sig['EMD_KOR_NM']=='법동')|(sig['EMD_KOR_NM']=='오정동')|(sig['EMD_KOR_NM']=='목상동')|(sig['EMD_KOR_NM']=='문평동')|(sig['EMD_KOR_NM']=='석봉동')|(sig['EMD_KOR_NM']=='관평동')|(sig['EMD_KOR_NM']=='봉산동')|(sig['EMD_KOR_NM']=='북대전')|(sig['EMD_KOR_NM']=='송강동')|(sig['EMD_KOR_NM']=='용산동')|(sig['EMD_KOR_NM']=='탑립동')]
sig_1
```

```python
sig_fin = sig_1[('30200000'<sig_1['EMD_CD']) & (sig_1['EMD_CD']<'30240000')]
sig_fin
```

```python
m = folium.Map(
    location=[36.4104119, 127.3845475],
    zoom_start=13, 
)

folium.GeoJson(
    sig_fin,
    name='지역구'
).add_to(m)

m
```
